**Q1.** Explain GET and POST methods.

**GET Method:**

Used to request data from a specified resource.

Data is sent in the URL as query parameters.

Data is visible in the browser's address bar.

Suitable for retrieving data like web pages, images, etc.

**POST Method:**

Used to submit data to be processed by a resource.

Data is sent in the request body.

Data is not visible in the URL or the address bar.

Suitable for actions that modify data, like form submissions or creating/updating records.

**Key Differences:**

GET is for data retrieval, while POST is for data submission.

GET sends data in the URL; POST sends it in the request body.

GET data is visible; POST data is not visible.

GET has limitations on data size; POST can handle larger data.

GET can be cached; POST is not cached by default.

POST is more secure for transmitting sensitive data.

**Q2.** Why is request used in Flask?

In Flask, the request object is used to handle incoming client requests made to the web server. It allows the Flask application to access and process data sent by the client in the HTTP request, such as form data, query parameters, request headers, and more. The request object is a part of Flask's request module and provides a convenient way to interact with the incoming request data.

**Data Retrieval:** It allows you to extract data from the incoming request, such as form data submitted by users, query parameters passed in the URL, or JSON data sent in the request body.

**Form Handling:** The request object enables you to access and process HTML form data submitted by users, making it easier to handle form submissions.

**Query Parameters:** It lets you access the query parameters in the URL, making it convenient to work with URL parameters.

**Cookies:** The request object provides access to cookies sent by the client, allowing you to read and manipulate cookie data.

**HTTP Headers:** It allows you to access the HTTP headers sent by the client, such as user-agent, content-type, etc.

**Session Handling:** The request object is also used in conjunction with the session object to handle client sessions and manage session data.

**File Uploads:** When handling file uploads from clients, the request object provides access to the uploaded files.

**Security:** Using the request object, you can validate and sanitize the incoming data, helping to prevent security vulnerabilities like cross-site scripting (XSS) attacks and SQL injection.

**Q3.** Why is redirect() used in Flask?


In Flask, the redirect() function is used to redirect a client's request to a different URL or route. When a web application receives a request, it may need to redirect the client to another page or route for various reasons. The redirect() function makes this process easy by sending a special HTTP response to the client, instructing it to visit the new URL.

**URL Redirection:** It allows you to redirect the client's request to a different URL. This is useful when a URL has changed, or when you want to direct the user to a different page or route after processing their request.

**Post-Submission Redirect:** After processing a form submission or any other user action that modifies data on the server, it is a best practice to redirect the user to a different page. This helps prevent accidental form resubmissions and ensures a smooth user experience.

**Route Abstraction:** Redirects provide a way to abstract routes and create cleaner URLs. For example, you can have multiple routes handle similar content but redirect them to a single canonical URL for SEO purposes.

**Security and Authentication:** Redirects are used during the authentication process. If a user tries to access a protected route without being authenticated, the server may redirect them to the login page.

**Handling Errors:** In certain cases, when an error occurs, you might want to redirect the user to a friendly error page or a page with troubleshooting instructions.

**Q4.** What are templates in Flask? Why is the render_template() function used?

**Templates in Flask:**

Templates in Flask are files that contain dynamic content to be rendered and sent as a response to the client.
They are written in HTML or other template languages like Jinja2, which allows embedding of dynamic data using placeholders or template variables.
Templates help separate the presentation logic from the application logic, promoting cleaner and more maintainable code.
Flask uses the Jinja2 template engine by default, which provides powerful template features like loops, conditionals, and template inheritance.

**render_template() Function:**

render_template() is a function in Flask that is used to render templates and produce HTML responses to be sent to the client.
It takes the name of the template file as an argument and processes it, filling in the dynamic data using template variables.
The function automatically looks for template files in the templates folder within the Flask application directory.
The data passed to render_template() can include variables, which are substituted into the template, making the rendered content dynamic and personalized for each user.
It enables dynamic web page generation by combining the template's structure with data from the backend, allowing customization based on user input or database queries.

**Q5.** Create a simple API. Use Postman to test it. Attach the screenshot of the output in the Jupyter Notebook.

```python
from flask import Flask, jsonify, request

app = Flask(__name__)


cricket_players = {
    7: "MS Dhoni",
    10: "Sachin Tendulkar",
    18: "Virat Kohli",
    45: "Rohit Sharma",
    99: "Sourav Ganguly"
}


@app.route('/api/cricket_players', methods=['GET'])
def get_cricket_players():
    return jsonify(cricket_players)

@app.route('/api/cricket_players/<int:jersey_number>', methods=['GET'])
def get_cricket_player(jersey_number):
    player_name = cricket_players.get(jersey_number)
    if player_name:
        return jsonify({str(jersey_number): player_name})
    return jsonify({"message": "Player not found"}), 404

if __name__ == '__main__':
    app.run()
```
